# 1. 필터링, 최솟값, 중앙값

### 1. 'f5' 칼럼이 0이 아닌 데이터(행)를 구하시오.

```python
import pandas as pd
df = pd.read_csv("type1_data1.csv")
df
```

```python
df_f5 = df[df['f5'] != 0]
df_f5
```

### 2. 앞에서 구한 데이터에 'views' 칼럼 결측치를 'views' 칼럼의 최솟값으로 채워주세요.

```python
df_f5['views'] = df_f5['views'].fillna(min(df_f5['views']))
df_f5['views'].isna()
```

### 3. 그리고 'views' 칼럼의 중앙값을 계산해 정수로 구하시오.

```python
round(df_f5['views'].median())
```

# 4. 값 변경, 정렬, 합계

### 1. 'views' 칼럼의 결측 데이터를 0으로 대체하시오.

```python
import pandas as pd
df = pd.read_csv("type1_data1.csv")
df['views'] = df['views'].fillna(0)
```

### 2. 'views' 칼럼에서 상위 10번째 값을 구하시오.

```python
views_10th = df.sort_values('views', ascending = False).head(10)['views'].iloc[-1]
views_10th
```

### 3. 'views' 칼럼에서 상위 10개의 값을 상위 10번째 값으로 대체하시오.

```python
df_views = df.sort_values('views', ascending = False)
df_views['views'][:10] = views_10th
df_views
```

### 4. 'views' 칼럼의 전체 합을 정수로 구하시오.

```python
int(df_views['views'].sum())
```

# 6. 필터링, 분산

# 7. 값 변경(연산), 필터링 절댓값

### 1. 모든 나이(age)에 1을 더하시오.

```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df['age'] = df['age'] + 1
```

### 2. 20대의 ‘views’ 평균과 30대의 ‘views’ 평균의 절댓값 차이를 구하시오.(반올림 후 소수 둘째 자리까지 계산)

```python
df_2 = df[df['age']//10 == 2]['views'].mean()
df_3 = df[df['age']//10 == 3]['views'].mean()
round(abs(df_2 - df_3),2)
```

# 11. 슬라이싱, 사분위수, 결측치 제거

### 1. 데이터에서 결측치가 있는 데이터(행)를 모두 제거하시오.

```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df = df.dropna().reset_index(drop = True)
```

### 2. 결측치가 제거된 데이터를 사용하여 앞에서부터 70% 데이터를 구하시오.(단, 데이터 70% 지점의 index가 소수점으로 계산될 경우 소수점 이하는 버림)

```python
end_point = int(len(df) * 0.7)
seventy = df.iloc[:end_point]
seventy
```

### 3. 앞에서 구한 70% 데이터중 'views' 칼럼의 3사분위수에서 1사분위수를 뺀 값을 정수로 구하시오.

```python
int(seventy['views'].quantile(0.75) - seventy['views'].quantile(0.25))
```

# 12. 결측치 처리, 최빈값, 데이터 개수

```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df.isna().sum().sort_values(ascending = False).index[:2]
```

### 2. 첫 번째로 결측치가 많은 칼럼에서 결측치가 있는 데이터(행)를 삭제하시오.

```python
df = df.dropna(subset = ['f1'])
```

# 14. 중복 데이터 제거, 값 변경, 데이터 개수

### 1. 중복 데이터를 제거하시오.

```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df = df.drop_duplicates()
```

### 2. ‘f3’ 칼럼의 결측치는 0, ‘silver’는 1, ‘gold’는 2, ‘vip’는 3으로 변환하시오.

```python
df['f3'] = df['f3'].fillna(0)
df['f3'] = df['f3'].replace('silver', 1)
df['f3'] = df['f3'].replace('gold', 2)
df['f3'] = df['f3'].replace('vip', 3)
```

### 3. 변환된 'f3' 칼럼의 총합을 정수형으로 구하시오.

```python
df['f3'].sum()
```
