# 16. 이상치, IQR

### 1. 'views' 칼럼의 1사분위수, 3사분위수 그리고 IQR을 계산하시오.

```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
quantile_1 = df['views'].quantile(0.25)
quantile_3 = df['views'].quantile(0.75)
print(quantile_1)
print(quantile_3)
iqr = quantile_3 - quantile_1
```

### 2. 이상치 조건에 맞는 데이터를 찾으시오.(이상치는 1사분위수 - (IQR * 1.5)보다 작은 값과 3사분위수 + (IQR * 1.5)보다 큰 값)

```python
df_filter = df[(df['views'] < quantile_1 - (1.5*iqr)) | (df['views'] > quantile_3 + (1.5 * iqr))]
```

### 3. 이상치 데이터의 'views' 칼럼 합을 정수로 구하시오.

```python
int(df_filter['views'].sum())
```

# 23. 시간 간의 차이 계산(분), 비율

### 1. 각 결제 종류별로 실제 도착 시간이 예상 도착 시간보다 늦은 주문의 비율을 계산하시오.

```python
import pandas as pd
df = pd.read_csv('delivery_time.csv')
df['late'] = df['실제도착시간'] > df['예상도착시간']
cnt = df.groupby(['결제종류'])['late'].count()
df_groupsum = df.groupby('결제종류').sum(numeric_only = True)
df_groupsum['ratio'] = df_groupsum['late'] / cnt
df_groupsum
```

```python
 round(df_groupsum['ratio'].max(),2)
```

# 25. 시간 간의 차이 계산(일)

### 1. 각 사용자별로 첫 주문과 마지막 주문 사이의 시간 간격을 일 단위로 계산하시오.

```python
import pandas as pd
df= pd.read_csv('delivery_time.csv')
df['주문시간'] = pd.to_datetime(df['주문시간'])
df_final = df.groupby(['user'])['주문시간'].max()
df_first = df.groupby(['user'])['주문시간'].min()
df_time = df_final - df_first
df_day = df_time.dt.total_seconds()/60/60/24
```

### 2. 시간 간격이 1일 이하인 사용자를 제외하고, 나머지 사용자들의 평균 시간 간격(일 단위)를 계산하시오.

```python
df_mean = df_day[df_day > 1].mean()
```

### 3. 평균 시간 간격보다 긴 시간 간격을 가진 사용자의 수를 정수로 구하시오.

```python
len(df_day[df_day>df_mean])
```

# 27. 시간 범위, 속도(km/h)

### 1. 점심시간(10시부터 13시 전까지)에 주문된 배달 데이터를 찾으시오.

```python
import pandas as pd
df = pd.read_csv('delivery_time.csv')
df['new'] = pd.to_datetime(df['주문시간']).dt.hour
filtered_df = df[df['new'].isin([10,11,12])]
filtered_df
```

### 2. 점심시간 주문 건 중 과속(평균 속도가 50km/h 이상) 하는 주문 수를 정수로 구하시오.
* 배달 시간 = 실제도착시간 - 주문시간
* 속도(km/h) = 거리(km) / 시간(h)

```python
filtered_df['실제도착시간'] = pd.to_datetime(filtered_df['실제도착시간'])
filtered_df['주문시간'] = pd.to_datetime(filtered_df['주문시간'])
filtered_df['배달시간'] = filtered_df['실제도착시간'] - filtered_df['주문시간']
filtered_df['배달시간'] = filtered_df['배달시간'].dt.total_seconds()/60/60
filtered_df['평균속도'] = filtered_df['거리'] / filtered_df['배달시간']
len(filtered_df[filtered_df['평균속도'] > 50])
```

```python
import pandas as pd
df = pd.read_csv('delivery_time.csv')

df['주문시간'] = pd.to_datetime(df['주문시간'])
df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])

df['시간'] = df['주문시간'].dt.hour
cond1 = df['시간'] >= 10
cond2 = df['시간'] < 13
df = df[cond1 & cond2]

df['배달시간'] = df['실제도착시간'] - df['주문시간']
df['배달시간'] = df['배달시간'].dt.total_seconds()/60/60
df['속도'] = df['거리'] / df['배달시간']
sum(df['속도']>=50)
```

# 29. 함수, 월별 집계

### 1. 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산하시오.
**[배달료 기준표]**
- 5km 미만 : 2000원
- 5km 이상 ~ 10km 미만 : 4000원
- 10km 이상 ~ 15km 미만 : 6000원
- 15km 이상 ~ 20km 미만 : 8000원

```python
import pandas as pd
df = pd.read_csv('delivery_time.csv')
def delivery_fee(distance) : 
    if distance < 5 : 
        return 2000
    elif distance < 10:
        return 4000
    elif distance < 15 : 
        return 6000
    else:
        return 8000
df['배달료'] = df['거리'].apply(delivery_fee)   
df
```

### 2. 월별로 배달료의 총합을 집계하시오.(연-월 별로 배달료의 총합을 집계하라는 문제이다.)

```python
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['month']= df['주문시간'].dt.to_period('M')
monthly = df.groupby(['month'])['배달료'].sum()
monthly
```

### 3. 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 구하시오. 

```python
print(monthly.idxmax())
print(monthly.loc[monthly.idxmax()])
```
